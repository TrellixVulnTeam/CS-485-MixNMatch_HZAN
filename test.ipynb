```python
from torchvision import transforms
from PIL import Image
import torchvision
import numpy as np

model = torchvision.models.segmentation.deeplabv3_resnet101(pretrained=True)
model.eval()
```

```python
import matplotlib.pyplot as plt 
```

```python
image = Image.open('kopek2.jpg')
```

```python
image
```

```python
x = transforms.ToTensor()(image).unsqueeze(0)
x = x[:, 0:3, :, :]
```

```python
x.shape
```

```python
out = model(x)
```

```python
smap = np.array(out['out'].detach())
aux = np.array(out['aux'].detach())
```

```python


for i in range(1, 22):
    plt.figure(figsize=[100, 100])
    plt.subplot(1, 22, i)
    img = smap[0, i - 1]
    plt.imshow(img)
    plt.show()
```

```python
im = np.array(x)
s = smap[0, 3] > 0
plt.imshow(s)
plt.axis('off')
plt.show()
```
